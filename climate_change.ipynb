```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import svm
import xgboost as xgb
```

```python
def Scores(yt,pred):
    MAE = mean_absolute_error(y_true=yt,y_pred=pred)
    MSE = mean_squared_error(y_true=yt,y_pred=pred)
    R2 = r2_score(y_true=yt,y_pred=pred)
    return {'MAE': MAE, 'MSE': MSE,'R2-score': R2}
```

```python
#Linear Regression
def lreg(xtn,xt,ytn,yt):
    lr = LinearRegression()
    lr.fit(xtn, ytn)
    pred = lr.predict(xt)
    return Scores(yt,pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures

def sreg(years,values):    
    # Reshape the data to fit the input requirements of LinearRegression
    X = years.reshape(-1, 1)

    # Create polynomial features of the sine function
    poly_features = PolynomialFeatures(degree=3)
    X_poly = poly_features.fit_transform(X)

    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_poly, values, test_size=0.2, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    pred = model.predict(X_test)
    return Scores(y_test,pred)
```

```python
#Decision Tree Regression
def dtreg(xtn,xt,ytn,yt):
    dtr = DecisionTreeRegressor(random_state=0)
    dtr.fit(xtn,ytn)
    pred = dtr.predict(xt)
    return Scores(yt,pred)
```

```python
#RandomForest Regress
def rfreg(xtn,xt,ytn,yt):
    rfr = RandomForestRegressor(n_estimators=100,random_state=0)
    rfr.fit(xtn,ytn)
    pred = rfr.predict(xt)
    return Scores(yt,pred)
```

```python
#SVM Regressor
def svreg(xtn,xt,ytn,yt):
    svr = svm.SVR(kernel='rbf', gamma='scale')
    svr.fit(xtn,ytn)
    pred = svr.predict(xt)
    return Scores(yt,pred)
```

```python
#KNN Regressor
def knnreg(xtn,xt,ytn,yt):
    #finding the best k value
    k_range = list(range(1, 31))
    param_grid = dict(n_neighbors=k_range)
    knn_reg = KNeighborsRegressor()
    grid = GridSearchCV(knn_reg, param_grid, cv=10, scoring='neg_mean_squared_error')
    grid.fit(xtn, ytn)
    best_k = grid.best_params_['n_neighbors']
    
    #regress
    knn_reg = KNeighborsRegressor(n_neighbors=best_k)
    knn_reg.fit(xtn, ytn)
    pred = knn_reg.predict(xt)
    
    return Scores(yt,pred)
```

```python
#XGBOOST 
def xgbreg(xtn,xt,ytn,yt):
    xgbreg = xgb.XGBRegressor()
    xgbreg.fit(xtn, ytn)
    pred = xgbreg.predict(xt)
    
    return Scores(yt,pred)
```

```python
#Deep Neural Network
def dnn(xtn,xt,ytn,yt):
    input_dim = xtn.shape[1]
    model = Sequential()
    model.add(Dense(12, activation='relu', input_dim=input_dim))
    model.add(Dense(12, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(xtn, ytn, epochs=10, batch_size=32, verbose=0)
    
    pred = model.predict(xt)
    return Scores(yt,pred)
```

```python

Area = pd.DataFrame()
Area['Area'] = df['Area'].unique()
Area['Code'] = df['Area Code (FAO)'].unique()
```

```python
from ipywidgets import interact,widgets
def Predict(x):
    
    # Filter data for specified country
    code = Area.loc[Area['Area'] == x, 'Code'].values[0]
    df = df_new[df_new['Area Code'] == code]
    
    # Prepare data for model training
    X = np.array(df['Year'].values).reshape(-1,1)
    Y = np.array(df['Value'].values).reshape(-1,1)
    
    split_data = train_test_split(X, Y, test_size = 0.3)
    benchmark = [
        lreg(*split_data),
        sreg(np.array(df['Year'].values),np.array(df['Value'].values)),
        dtreg(*split_data),
        rfreg(*split_data),
        svreg(*split_data),
        knnreg(*split_data),
        xgbreg(*split_data),
        dnn(*split_data)
    ]
    labels = ['LinearRegression','Sine Regression','Decision Tree','Random Forest','Support Vector Machines','K-nearest Neighbors','XGBoost','Deep Neural Neutral']
    col = ['MAE','MSE','R2-score']
    benchmark_table = pd.DataFrame(benchmark,index=labels,columns=col)
    table_widget = widgets.HTML(benchmark_table.to_html())
    display(table_widget)
    
    # Find the model with the best overall performance
    overall_scores = benchmark_table.mean(axis=1)
    best_model = overall_scores.idxmin()
    
    ax = benchmark_table.plot(kind='bar', rot=0, figsize=(10, 6))
    ax.set_title('Comparison of Model Scores')
    ax.set_xlabel('Model')
    ax.set_ylabel('Score')
    
    # Highlight the bars with the best overall performance
    highlight_label = 'Best Model'
    best_model_index = labels.index(best_model)
    ax.annotate(highlight_label, xy=(best_model_index, overall_scores[best_model]),
                xytext=(0, 3), textcoords='offset points',
                ha='center', va='bottom')
    # Adjust spacing between labels
    plt.xticks(range(len(labels)), labels, rotation=45, ha='right')
    plt.subplots_adjust(bottom=0.2)  # Adjust the bottom margin as needed

    # Display the plot
    plt.show()
```

```python
interact(Predict, x =widgets.Dropdown(options = Area['Area'], description = 'Country Name: '))
```
